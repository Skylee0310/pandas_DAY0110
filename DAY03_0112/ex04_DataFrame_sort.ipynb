- [실습] 다양한 정렬 <hr>

```python
import numpy as np
from math import nan                #nan : Not a Number 빈칸 의미
data={'col1':['A','A','B',nan,'D','C'],
      'col2':[2,1,9,8,7,4],
      'col3':[0,1,9,4,2,3],
      'col4':['a','B','c','D','e','F'],}
dataDF=pd.DataFrame(data)
dataDF
```

- col1 컬럼을 기준으로 오름차순 정렬

```python
dataDF.sort_values(by='col1')# Nan 즉, 빈칸 데이터(결측치欠測値) 값의 위치 : 제일 마지막
```

```python
# Nan 즉, 빈칸 데이터(결측치欠測値) 값의 위치 설정 => na_position 파라미터
dataDF.sort_values(by='col1', na_position='first')
```

```python
# col4 컬럼을 기준으로 오름차순 정렬
dataDF.sort_values(by='col4')
```

```python
dataDF.sort_values(by='col4', key=lambda col : col.str.upper()) #이해 못했음...★!
```
